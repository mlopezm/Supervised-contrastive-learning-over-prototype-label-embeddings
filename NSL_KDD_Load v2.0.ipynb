## List of features for test and training data

Description of features.

The feature to be predicted is the class feature. This feature shows if the connection has been normal or has presented some kind of attack (anomaly)

The class feature has different number of label names for training and test data,with more label names for test data.

That means that there are anomalies that has not been presented on training. This is important as it provides the opportunity to extend prediction beyond training information 

The difficulty feature shows the number of predictors that managed to predict correctly that connection. This was done as part of an experiment using 21 predictors in combination, trying to predict the class feature. A higher number here shows that the connection was easier to predict.

## Table with counts of class vs difficulty

Table of frequency counts for class names vs difficulty level, for training data

```python
# table with counts of class vs difficulty
cxClassDiff = pd.crosstab(dataTrain['class'],dataTrain['difficulty'], margins = True)
# marginals in percentage
cxClassDiff['Perc'] = [i/(0.5*sum(cxClassDiff['All'])) for i in cxClassDiff['All'] ]
cxClassDiff.ix[-1,-1] = 0
cxClassDiff.loc['Perc'] = [i/(0.5*sum(cxClassDiff.loc['All'])) for i in cxClassDiff.loc['All'] ]

cxClassDiff
```


##Difference in labels between test data and training time

We provide the frequency of new labels in test data that were not present at training time and the list of these new values.

We provide as well the labels not in the intersection of the two sets of labels (train and test)

```python
trainUniqVals = dataTrain['class'].unique()
testUniqVals = dataTest['class'].unique()
diffUniqVals = set(testUniqVals)-set(trainUniqVals)
InterUniqVals = set(testUniqVals).intersection(set(trainUniqVals))

newTestTag = dataTest['class'].apply(lambda i: 1 if i in diffUniqVals else 0)
print('Frequency of new labels in test data that were not present at training time:')
print(np.mean(newTestTag))
print('\nList of labels not in the intersection between the sets at test and training time:')
print(set(testUniqVals).union(trainUniqVals)-InterUniqVals)
print('\nNumber of labels not in the intersection between the sets at test and training time:')
print(len(set(testUniqVals).union(trainUniqVals)-InterUniqVals))
print('\nList of labels in test but not in training:')
print(set(testUniqVals)-set(trainUniqVals))
print('\nNumber of labels in test but not in training:')
print(len(set(testUniqVals)-set(trainUniqVals)))
print('\nList of labels in training but not in test:')
print(set(trainUniqVals)-set(testUniqVals))
print('\nNumber of labels in training but not in test:')
print(len(set(trainUniqVals)-set(testUniqVals)))
print('\nList of labels  in the intersection between the sets at test and training time:')
print(InterUniqVals)
print('\nNumber of labels in the intersection between training and test sets')
print(len(InterUniqVals))
```
