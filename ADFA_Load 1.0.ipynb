
## Load train and test data

We load the train and test data and show the head of the Train and Test files


```python
#load data
# train and test names are wrong
dataTest = pd.read_csv("UNSW_NB15_training-set.csv", header = 0)
dataTrain = pd.read_csv("UNSW_NB15_testing-set.csv", header = 0)
print(dataTrain.shape)
print(dataTest.shape)
dataTrain.head(5)
```

```python
# limit to categorical data using df.select_dtypes()

#Xctr and Xcte are the categorical parts
Xctr = dataTrain.select_dtypes(include=[object])
Xcte = dataTest.select_dtypes(include=[object])
Xcte.head(10)
```

```python
Xctr.shape
```

```python
Xcte.shape
```

```python
# import preprocessing from sklearn
from sklearn import preprocessing

# encode labels with value between 0 and n_classes-1.
le1 = preprocessing.LabelBinarizer()
le2 = preprocessing.LabelBinarizer()
le3 = preprocessing.LabelBinarizer()
le4 = preprocessing.LabelBinarizer()

Xctotal = Xctr.append(Xcte, ignore_index=True)
print(Xctotal.shape)
Xctotal1 = le1.fit_transform(Xctotal["proto"])
Xctotal2 = le2.fit_transform(Xctotal["service"])
Xctotal3 = le3.fit_transform(Xctotal["state"])
Xctotal4 = le4.fit_transform(Xctotal["attack_cat"])

# Train
Xctr1 = Xctotal1[0:Xctr.shape[0]]
print("valores proto: {}".format(Xctr1.shape))

Xctr2 =  Xctotal2[0:Xctr.shape[0]]
print("valores service: {}".format(Xctr2.shape))
Xctr3 =  Xctotal3[0:Xctr.shape[0]]
print("valores state: {}".format(Xctr3.shape))
Xctr4 =  Xctotal4[0:Xctr.shape[0]]
print("valores attack_cat: {}".format(Xctr4.shape))
```

```python
# Test
# encode labels with value between 0 and n_classes-1.
Xcte1 = Xctotal1[Xctr.shape[0]:]
print("valores proto: {}".format(Xcte1.shape))
Xcte2 =  Xctotal2[Xctr.shape[0]:]
print("valores service: {}".format(Xcte2.shape))
Xcte3 =  Xctotal3[Xctr.shape[0]:]
print("valores state: {}".format(Xcte3.shape))
Xcte4 =  Xctotal4[Xctr.shape[0]:]
print("valores attack_cat: {}".format(Xcte4.shape))
```

```python
data_total = dataTrain.append(dataTest, ignore_index=True)
data_total_continous = data_total.select_dtypes(exclude=[object]).drop(['id','label'], axis=1)
data_total_continous.shape
```

```python
# scale the values in X
# columns 3 to 24 with log and min-max and rest only min-max
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
for n in data_total_continous.columns.values[0:27]:
    data_total_continous[n] = mm.fit_transform(np.log(data_total_continous[n].values.reshape(-1, 1)+1))
for n in data_total_continous.columns.values[27:]:
    data_total_continous[n] = mm.fit_transform(data_total_continous[n].values.reshape(-1, 1))
```

# Build Xtrain and Ytrain

```python
# Build Xtrain
Xtrain = data_total_continous.iloc[0:Xctr.shape[0],:]
print(Xtrain.shape)

Xtrain = pd.concat([Xtrain, 
                      pd.DataFrame(Xctr1),
                      pd.DataFrame(Xctr2),
                      pd.DataFrame(Xctr3)], axis=1)
print(Xtrain.shape)
```

```python
# Build Ytrain
Ytrain_label = dataTrain['label']
Ytrain_attack = Xctr4
print(Ytrain_label.shape)
print(Ytrain_attack.shape)
Ytrain_attack_char = dataTrain['attack_cat']
print(Ytrain_attack_char.shape)
```

# Build Xtest and Ytest

```python
Xtest = data_total_continous.iloc[Xctr.shape[0]:,:]
Xtest.reset_index(inplace = True, drop=True)
print(Xtest.shape)

Xtest = pd.concat([Xtest, 
                      pd.DataFrame(Xcte1),
                      pd.DataFrame(Xcte2),
                      pd.DataFrame(Xcte3)], axis=1 )
print(Xtest.shape)
```

```python
# Build Ytrain
Ytest_label = dataTest['label']
Ytest_attack = Xcte4
print(Ytest_label.shape)
print(Ytest_attack.shape)
Ytest_attack_char = dataTest['attack_cat']
print(Ytest_attack_char.shape)
```
